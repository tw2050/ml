```python
from skimage import io
```

```python
predictor_path = 'F:/dlib/shape_predictor_68_face_landmarks.dat'
face_rec_model_path = 'F:/dlib/dlib_face_recognition_resnet_model_v1.dat'
face_floder_path = 'F:/dlib/src_img'
img_path = 'F:/dlib/src_img/guanzl5.jpg'
```

```python
## 人脸数量检测器
detector = dlib.get_frontal_face_detector()
```

```python
## 人脸关键点检测器（根据人脸提取68个关键点）
sp = dlib.shape_predictor(predictor_path)
```

```python
## 人脸识别模型（根据人脸关键点计算， 得到一个128维的数据）
facerec = dlib.face_recognition_model_v1(face_rec_model_path)
```

```python
descriptors = []
for f in glob.glob(os.path.join(face_floder_path, "*1.jpg")):
    print("procession file: {}".format(f))
    img = io.imread(f)
    
    dets = detector(img, 1)
    print("number of faces detected: {}".format(len(dets)))
    for k, d in enumerate(dets):
        shape = sp(img, d)
        face_descriptor = facerec.compute_face_descriptor(img, shape)
        v = np.array(face_descriptor)
        descriptors.append(v)
        print("v: {}".format(v))
```

```python
img = io.imread(img_path)
print("img_path:{}".format(img_path))
ets = detector(img, 1)
dist = []
for k, d in enumerate(ets):
    shape = sp(img, d)
    face_descriptor = facerec.compute_face_descriptor(img, shape)
    d_test = np.array(face_descriptor) 
    print("d_test:{}".format(d_test)) 
        
    # 计算欧式距离
    for i in descriptors:
        dist_ = np.linalg.norm(i-d_test)
        dist.append(dist_)

dist
```

```python
candidate = ['fanbb','guanzl','liangjh','liudh','liuss','tangw']
```

```python
c_d = dict(zip(candidate,dist))
cd_sorted = sorted(c_d.items(), key=lambda d:d[1])
print('\n The person is: %s' % ( cd_sorted[0][0] )  )  
dlib.hit_enter_to_continue()
```
