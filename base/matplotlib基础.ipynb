```python
x1 = np.linspace(0,10, 100)
siny = np.sin(x1)
cony = np.cos(x1)
plt.figure()
plt.plot(x1,siny,color = "red", linestyle = "--")
plt.plot(x1,cosy)
plt.show()
```

```python
x1 = np.linspace(0,10, 100)
siny = np.sin(x1)
cony = np.cos(x1)
plt.figure()
plt.plot(x1,siny,color = "red", linestyle = "--")
plt.plot(x1,cosy)
plt.xlim(-2, 12)
plt.ylim(0, 2)
plt.show()
```

```python
x1 = np.linspace(0,10, 100)
siny = np.sin(x1)
cony = np.cos(x1)
plt.figure()
plt.plot(x1,siny,color = "red", linestyle = "--")
plt.plot(x1,cosy)
plt.axis([-2,12,0,2])
plt.show()
```

```python
x1 = np.linspace(0,10, 100)
siny = np.sin(x1)
cony = np.cos(x1)
plt.figure()
plt.plot(x1,siny,color = "red", linestyle = "--")
plt.plot(x1,cosy)
plt.xlabel("x axis")
plt.ylabel("y val")
plt.show()
```

```python
x1 = np.linspace(0,10, 100)
siny = np.sin(x1)
cony = np.cos(x1)
plt.figure()
plt.plot(x1,siny,color = "red", linestyle = "--", label = "y = sinx")
plt.plot(x1,cosy, label = "y = cosx")
plt.xlabel("x axis")
plt.ylabel("y val")
plt.legend()
plt.title("sin & cos")
plt.show()
```

```python
x1 = np.linspace(0,10, 100)
siny = np.sin(x1)
cony = np.cos(x1)
plt.figure()
plt.scatter(x1,siny)
plt.scatter(x1,cosy)
plt.show()
```
