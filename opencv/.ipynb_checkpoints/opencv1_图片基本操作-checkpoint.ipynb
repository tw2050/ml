```python
### 图片的打开， 复制， 保存

import cv2
import numpy as np
```

```python
## 图片读取-显示

img_path = 'F:/opencv/img/cat2.jpg'
img = cv2.imread(img_path)
cv2.namedWindow('img-show', cv2.WINDOW_AUTOSIZE)
cv2.imshow('img-show', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
### 图片复制

temp_img = np.zeros(img.shape, np.uint8)
temp_img = img.copy()
```

```python
cv2.namedWindow('img-copy', cv2.WINDOW_NORMAL)
cv2.imshow('img-copy',temp_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
### 灰度
temp_img1 = np.zeros(img.shape, np.uint8)
temp_img1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2.imshow('img-gray',temp_img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
### 保存图片
cv2.imwrite('f:/opencv/img/cat_gray.jpg', temp_img1)
```

```python
### 保存png 第三个参数为压缩等级0-9  越高压缩越厉害
cv2.imwrite('f:/opencv/img/cat_gray9.png', temp_img1, [int(cv2.IMWRITE_PNG_COMPRESSION), 9])
```

```python
cv2.imwrite('f:/opencv/img/cat_gray0.png', temp_img1, [int(cv2.IMWRITE_PNG_COMPRESSION), 0])
```
