```python
stu_data['status'] = stu_data.apply(lambda x : 'Fail' if x['pass_math'] == 'Fail' or 
                           x['pass_read'] == 'Fail' or x['pass_write'] == 'Fail'
                           else 'pass', axis = 1)   ## 全科通过则Pass 否则 Fail
stu_data['status'].value_counts(dropna = False).plot.pie(colors = ['grey', 'crimson'])  # 饼图
plt.title('overall results', fontweight = 30, fontsize = 20)
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

```python
# Assigning grades to the grades according to the following criteria :
# 0 - 60 marks : grade D
# 60 - 70 marks : grade C
# 70 - 80 marks : grade B
# 80 - 90 marks : grade A
# 90 - 100 marks : grade S

def getgrade(percentage, status):
  if status == 'Fail':
    return 'D'
  if(percentage >= 90):
    return 'S'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  else :
    return 'D'
```

```python
stu_data['grades'] = stu_data.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )
stu_data['grades'].value_counts()
```

```python
labels = ['Grade S', 'Grade A', 'Grade B', 'Grade C', 'Grade D']
sizes = [58, 156, 253, 136, 397]
colors = ['yellow', 'gold', 'lightskyblue', 'lightcoral', 'pink']
explode = (0.0001, 0.0001, 0.0001, 0.0001, 0.0001, 0.0001)

patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=150)
plt.legend(patches, labels)
plt.title('Distribution of Grades among Students', fontweight = 30, fontsize = 20)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 9)
x = pd.crosstab(stu_data['parental level of education'], stu_data['grades'])  ## parental level of education 与grades 行列关系
color = plt.cm.hot(np.linspace(0, 1, 8))  ## 颜色渐变
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True, color = color)  # 柱状图
plt.title("Parental Education vs Student's Grades", fontweight = 30, fontsize = 20)
plt.show()
```

### Data Preparation

```python
stu_data.shape
```

```python
x = stu_data.iloc[:, :14]  # 所有行， 前14列设置为x
y = stu_data.iloc[:, 14]  # 所有行， 最后一列设置为y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.25, random_state = 45)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
## 归一化， 常用的有minmaxscaler -- 最值归一法，还有standardscaler , 后者更常用
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)
```

## 模型
## logistic regression  逻辑回归

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(train_x, train_y)
predict_y = model.predict(test_x)
```

```python
print("Training Accuracy :", model.score(train_x, train_y))
print("Testing Accuracy :", model.score(test_x, test_y))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
## 混淆矩阵， 用于查看分类效果
cm = confusion_matrix(test_y, predict_y)
```

```python
plt.rcParams['figure.figsize'] = (8, 8)
sns.heatmap(cm, annot = True, cmap = 'Greens')
plt.title('Confusion Matrix for Logistic Regression', fontweight = 30, fontsize = 20)
plt.show()
```

### 随机森林

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(train_x, train_y)
predict_y = model.predict(test_x)
```

```python
print("Training Accuracy :", model.score(train_x, train_y))
print("Testing Accuracy :", model.score(test_x, test_y))
```

```python
cm = confusion_matrix(test_y, predict_y)
# printing the confusion matrix
plt.rcParams['figure.figsize'] = (8, 8)
sns.heatmap(cm, annot = True, cmap = 'Reds')
plt.title('Confusion Matrix for Random Forest', fontweight = 30, fontsize = 20)
plt.show()
```
